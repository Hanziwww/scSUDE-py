## scVI

```python
adata_path = os.path.join("tmp", "lung_atlas_preprocessed.h5ad")

adata = sc.read(
    adata_path,
    backup_url="https://figshare.com/ndownloader/files/52859312",
)
adata
```

```python
scvi.model.SCVI.setup_anndata(adata, layer="counts", batch_key="batch")
```

```python
model = scvi.model.SCVI(adata, n_layers=2, n_latent=30, gene_likelihood="nb")
```

```python
model.train()
```

```python
SCVI_LATENT_KEY = "X_scVI"
adata.obsm[SCVI_LATENT_KEY] = model.get_latent_representation()
```

```python
sc.pp.neighbors(adata, use_rep=SCVI_LATENT_KEY)
sc.tl.umap(adata)
sc.tl.leiden(adata)
```

```python
import numpy as np

SCVI_LAYER = "scvi_denoised"
gchunk = 1024
nsamples = 25

den_all = np.empty((adata.n_obs, adata.n_vars), dtype=np.float16)

for s in range(0, adata.n_vars, gchunk):
    sub_cols  = np.arange(s, min(s + gchunk, adata.n_vars))
    sub_names = adata.var_names[sub_cols]

    den = model.get_normalized_expression(
        adata=adata,
        gene_list=sub_names,
        n_samples=nsamples,
        library_size=50*1e4,
        return_mean=True,
        return_numpy=True,
        batch_size=1024,
    ).astype(np.float16)

    den_all[:, sub_cols] = den

adata.layers[SCVI_LAYER] = den_all
```

```python
sc.pl.umap(
    adata,
    color=["batch", "cell_type", "leiden"],
    frameon=False,
    ncols=1,
    save="_UMAP.pdf"
)
```

```python
adata.write("sude_test.h5ad")
```
